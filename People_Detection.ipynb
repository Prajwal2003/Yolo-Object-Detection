# **Performance** **Metrics**

```python
from ultralytics import YOLO
import torch
from torchmetrics.detection.mean_ap import MeanAveragePrecision
import cv2
import matplotlib.pyplot as plt

model = YOLO('/content/runs/detect/train/weights/best.pt')

results = model('/content/drive/MyDrive/Human-dataset/images/val', save=False, save_txt=False, conf=0.01)

mAP50 = MeanAveragePrecision(iou_type='bbox')

predictions = []
ground_truth = []

for result in results:
    if result.boxes.xyxy.numel() > 0:
        boxes = result.boxes.xyxy.cpu()
        scores = result.boxes.conf.cpu()
        labels = result.boxes.cls.cpu()
        predictions.append({
            "boxes": boxes,
            "scores": scores,
            "labels": labels.int()
        })
        ground_truth.append({
            "boxes": torch.tensor([[0, 0, 10, 10]]),
            "labels": torch.tensor([0])
        })

if predictions:
    mAP50.update(predictions, ground_truth)

    best_result = max(
        (r for r in results if r.boxes.conf.numel() > 0),
        key=lambda r: r.boxes.conf.max().item()
    )

    max_confidence_idx = best_result.boxes.conf.argmax().item()
    box = best_result.boxes.xyxy[max_confidence_idx].cpu().tolist()
    score = best_result.boxes.conf[max_confidence_idx].cpu().item()
    label = best_result.boxes.cls[max_confidence_idx].cpu().item()

    image_path = best_result.path
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    x1, y1, x2, y2 = map(int, box)
    cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)
    cv2.putText(image, f'{label}: {score:.2f}', (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    plt.figure(figsize=(10, 10))
    plt.imshow(image)
    plt.axis('off')
    plt.show()

    print(f'mAP@50: {mAP50.compute()}')

else:
    print("No predictions with bounding boxes found.")
```
